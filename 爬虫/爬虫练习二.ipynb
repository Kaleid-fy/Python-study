```python
def getdetail(soup):
    infor = {}
    infor['title'] = soup.select('.book-breintro')[0].select('h3')[0].text
    infor['book public'] = soup.select('.book-breintro')[0].select('.book-public')[0].text.strip().split('/')
    infor['original/current price'] = soup.select('.book-breintro')[0].select('.book-price')[0].text.strip().split('\n') 
    infor['total score'] = soup.select('.score-left .total-score')[0].text
    infor['comment number'] = soup.select('.score-left .comment-number')[0].text
    return infor
    #字典默认按照字母顺序排列
```

```python
urls = ['https://market.douban.com/book/?utm_campaign=book_nav_freyr&utm_source=douban&utm_medium=pc_web&page={}&page_num=18&'.format(str(i)) for i in range(1,3)]
print(urls)
bookary = []
for url in urls:
    res = requests.get(url)
    soup0 = BeautifulSoup(res.text, 'html.parser')
    link = []
    for bb in soup0.select('.books-wrap .book-item'):
        link.append(bb.select('a')[0]['href']) 
    for booklink in link:
        res = requests.get(booklink)
        soup = BeautifulSoup(res.text, 'html.parser')
        print(booklink)
        bookary.append(getdetail(soup))
```
