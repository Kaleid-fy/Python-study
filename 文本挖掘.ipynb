```python
import pandas as pd
import jieba
import re
```

```python
mescon_all = pd.read_csv('whole.csv',header=None,encoding='UTF-8',error_bad_lines=False)
```

```python
mescon_all
```

```python
outfile = open('result.csv','w')
#逐条取出短信Label与内容进行处理 并将处理结果写入文件中
for i in range(len(mescon_all)):
	#80万条短信内容存于mescon_all[2]数组中
    mescon_single = mescon_all[2][i] 
	#80万条短信Label存于mescon_all[1]数组中
    me_cate = mescon_all[1][i] 
    outstr = ''
	#正则表达式选取 除中文和英文以外的符号等全部删去
    temp = re.sub(u'[^\u4e00-\u9fa5A-Za-z]','',mescon_single) 
	#再将筛选好的数据用jieba进行分词处理
    ms_cut = list(jieba.cut(temp,cut_all=False)) 
	
	#分好词的结果写入文件中
    for word in ms_cut:
        if word != ' ':
             outstr += word+' '
    outfile.write((str(me_cate)+','+outstr).encode('UTF-8').decode('UTF-8')+'\n')

outfile.close()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import HashingVectorizer
```

```python
vector = HashingVectorizer(n_features=100)
temp = vector.transform(mescon_all[1][0:10]).todense()
```

```python
listtodel = []
#此处目的是将内容为空的短信删除，只保留可以转成特征向量的短信
listtodel = []
for i,line in enumerate(mescon_all[1]):
    if type(line)!=str:
        listtodel.append(i)
mescon_all = mescon_all.drop(listtodel) 

outfile = open('features.txt','w')

#采用HashingVectorizer方法进行特征向量转化
vector = HashingVectorizer(n_features=100)
temp = vector.transform(mescon_all[1]).todense()

#用drop方法筛选内容时，解决索引不对应情况
x = [[i,j] for i,j in enumerate(mescon_all[0])]

temp = temp.tolist()
for i,line in enumerate(temp):
    outstr = ''
    for word in line:
        outstr += str(word+1)
        outstr += ' ' 
    outfile.write((str(mescon_all[0][x[i][1]])+','+outstr).encode('utf-8').decode('UTF-8')+'\n')
    
outfile.close()
```
